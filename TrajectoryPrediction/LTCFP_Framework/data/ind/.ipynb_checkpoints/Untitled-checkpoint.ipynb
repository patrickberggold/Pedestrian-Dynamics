```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
```

```python
!ls '00'
```

```python
for folder in os.listdir('.'):
    try:
        np_image = np.load(f'{folder}/scene_mask.npy')
        np_image = np_image.argmax(axis=-1)
        cv2.imwrite(f"{folder}/pred_mask.png", np_image)
        os.remove(f'{folder}/scene_mask.npy')
    except:
        print(folder)
```
